```python
import numpy as np
import pandas as pd
#used to store and echange complex data: allows  you to save python objects and then later load them back into memory
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#import data
data = pd.read_csv("Loan_Default.csv", index_col=0)
```

```python
#look at the data and categories
data.head()
```

```python
#making all variable names lower case for convience
data.columns = data.columns.str.lower()
data.head()
```

```python
#look at infor to see columns, couts, and type
data.info()
```

```python
#look at missing values
data.isnull().sum()
```

```python
#look at the unique factors of status
data['status'].unique()
```

```python
#look at the counts of each category
data['status'].value_counts(dropna=False)
```

```python
#look at the count of each category for credit score
#there are two many variables to look at them individualy compared to status
#need to bin the variables to analyze them
data['credit_score'].value_counts(dropna=False)
```

```python
#use groupby and bins to put the credit_score varibles in bins by credit_score range (poor, fair, good, very good, exceptional)
groups = data.groupby(['status', pd.cut(data.credit_score, [300, 579, 669, 739, 799, 850])])

#display bin count grouped by credit score
groups.size().unstack()
```

```python
# plot the result
groups.size().unstack().plot.bar()
plt.xticks(rotation=45)
plt.title("Count of Credit Score Defaulted or Not Defaulted")
plt.show()
```

```python
#put credit_score variable into a bin by credit_score category (poor, fair, good, very good, exceptional).

credit = data['credit_score']
bins = [300, 579, 669, 739, 799, 850]
cats = pd.cut(credit, bins)
cats
```

```python
# countplot of credit_score categories
plt.figure(figsize = (6,6))
sns.countplot(x = cats, data = data)
plt.title("Count of Credit Score within Categories")
plt.show()
```
